## CelebA 1 examples

```python
import numpy as np
import torch
import torchvision
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
import src.mdata
```

```python
torch.manual_seed(1)
np.random.seed(1)
dataloaders = src.mdata.celeba.celeba_perturb(perturb = 'CelebA_1',
                                  image_path = '../data/CelebA/img_align_celeba',
                                  attr_path = '../data/CelebA/attr/list_attr_celeba.txt')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(12,4))
for i, j in zip([2,1,0,3], [0,1,2,3]):
    for x, y in dataloaders[j]:
        img = torchvision.utils.make_grid(torch.tensor(x[:16]),
                                      nrow = 4)
        ax[i].imshow(np.transpose(img.numpy(), (1,2,0)))
        ax[i].axis('off')
        if i < 3:
            ax[i].set_title(f'Source domain {i+1}')
        else:
            ax[i].set_title(f'Target domain')
        break
plt.tight_layout(pad=1.2)
plt.savefig('../results/CelebA.png', dpi=200)
plt.show()
```
