## DomainNet examples

```python
import numpy as np
import torch
import torchvision
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
import src.mdata
```

```python
torch.manual_seed(1)
np.random.seed(1)
dataloaders = src.mdata.domainnet.domainNet_loaders(
    directory='../data/domainNet',
    divs=['Furniture', 'Mammal'],
    batch_size=16
    )
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))
axes = axes.flatten() 
for i, j in zip(range(6), [0, 1, 2, 3, 5, 4]):
    for x, y in dataloaders[j]:
        img = torchvision.utils.make_grid(torch.tensor(x[:16]), nrow=4)
        axes[i].imshow(np.transpose(img.numpy(), (1, 2, 0)))
        axes[i].axis('off')
        if i < 5:
            axes[i].set_title(f'Source domain {i+1}')
        else:
            axes[i].set_title('Target domain')
        break

plt.tight_layout(pad=1.2)
plt.savefig('../results/DomainNet.png', dpi=200)
plt.show()
```
